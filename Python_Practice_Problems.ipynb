**Control Flow**

```python
Age = 18

if Age >= 18 :
    print("Eligible to VOTE")
```

```python
Num = int(input("Enter Number : "))

if Num > 0 :
    print("Positive")
else:
    print("Negative")
```

```python
T = 14

if T % 2 == 0 :
    print("Even")
else :
    print("Odd")
```

```python
Marks = 76

if Marks >= 80 :
    print("Grade : A+")
elif Marks >= 70 :
    print("Grade : A")
elif Marks >= 60 :
    print("Grade : A-")
elif Marks >= 50 :
    print("Grade : B")
else :
    print("Fail")
```

```python
Signal = input("Enter Traffic Signal Color : ")

if Signal == "Red" :
    print("Stop")
elif Signal == "Yellow" :
    print("Get Ready")
elif Signal == "Green" :
    print("Go")
else :
    print("Invalid Signal")
```
