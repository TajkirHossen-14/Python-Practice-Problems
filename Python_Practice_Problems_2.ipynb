### **Python Data Structures**

**- Tuples -**

```python
T1 = (1, 2, 3, 4, 5)
T2 = (8, 14.21, "Bangladesh", True)

print(T1)
print(T2)
```

```python
T1 = (7)          # Integer
T2 = (7, )        # Tuple

print(type(T1))
print(type(T2))
```

```python
T = (10, 20, 30, 40)
print(T[0])
print(T[-1])
```

```python
T = ("A", "B", "C")
print(len(T))
```

```python
T = (100, 200, 300)

for Number in T:
    print(Number)
```

```python
T = (10, 20, 30, 40)

for j in range(len(T)):
  print(j, T[j])

print("----")

k = 0

while k < len(T):
  print(T[k])
  k += 1
```

```python
# Tuple Slicing                         Tuple[Start : End : Step]

T = (1, 2, 3, 4, 5, 6, 7, 8)            # Print from starting index to just before the last index.

print(f"1.  {T[:]}")
print(f"2.  {T[1:7]}")
print(f"3.  {T[1:8:3]}")
print(f"4.  {T[:4]}")
print(f"5.  {T[4:]}")
print(f"6.  {T[::2]}")
print(f"7.  {T[::-1]}")
print(f"8.  {T[-3:]}")
print(f"9.  {T[-4:-1]}")
print(f"10. {T[-5:4]}")
print(f"11. {T[5:-4]}")
print(f"12. {T[-1:-8:-3]}")
print(f"13. {T[1:100]}")
```

```python
T1 = (1, 2, 3)
T2 = (4, 5, 6)

T3 = T1 + T2
print(T3)
```

```python
T = (10, 25, 50)

print(min(T))
print(max(T))
```

```python
T = (505, 506, 507, 508)

print(sum(T))
```

```python
T = (1, 3, 5, 7, 3, 9, 3, 11)

print(T.count(3))
```

```python
T = (10, 30, 60, 90)

print(T.index(60))
```

```python
T = (6, 2, 4)

print(sorted(T))         # Return as List

print(tuple(sorted(T)))
```

```python
# Tuple Packing
Tuple = "KTH", 22, "Student"

print(Tuple)

print("\n")

# Tuple Unpacking
Name, Age, Profession = Tuple

print(Name)
print(Age)
print(Profession)
```

**- Python Dictionaries -**

```python
# Create Dictionary :      Dict = {"Key1" : "Value1", "Key2" : "Value2", "Key3" : "Value3",...}

Student = {
    "Name": "Tajkir",
    "Age": 22,
    "City": "Chandpur"
}

print(Student)
```

```python
Student = dict(Name = "Tajkir", Age = 22, City = "Dhaka")           # Using  dict()  function
print(Student)
```

```python
print(f"1.  {Student.keys()}")                  # Print keys
print(f"2.  {Student.values()}")                # Print values
print(f"3.  {Student.items()}")                 # Print key-value pairs
```

```python
print(Student["Name"])                 # Access values by key
print(Student["Age"])
print(Student["City"])
```

```python
Student = {"Name": "Tajkir", "Age": 22, "City": "Chandpur"}

print(Student.get("Name"))             # Access values using  get()  function
print(Student.get("Age"))
print(Student.get("City"))
```

```python
Student["CGPA"] = 3.5                  # Add new key-value
print(Student)
```

```python
Student["Age"] = 23                    # Update value
print(Student)
```

```python
Student = {"Name": "Tajkir", "Age": 22, "City": "Chandpur", "CGPA": 3.5}

Student.update({"Age" : 23, "CGPA": 3.6})             # Using  update()  function

print(Student)
```

```python
Student = {"Name": "Tajkir", "Age": 22, "City": "Chandpur", "CGPA": 3.5}

del Student["CGPA"]                    # Delete a key

print(Student)
```

```python
Student = {"Name": "Tajkir", "Age": 22, "City": "Chandpur"}

Student.pop("Age")                     # Using  pop()  function

print(Student)
```

```python
Student = {"Name": "Tajkir", "Age": 22, "City": "Chandpur"}

Student.popitem()                      # Delete last item

print(Student)
```

```python
Student.clear()                        # Delete all items

print(Student)
```

```python
Student = {"Name": "Tajkir",  "Age": 22,  "City": "Chandpur"}

print(len(Student))
```

```python
Student = {"Name": "Tajkir", "Age": 22, "CGPA": 3.5}

if 3.5 in Student.values() :
    print("Value Exists")
else :
    print("Value Does Not Exist")
```

```python
Student = {"Name": "Tajkir", "Age": 22}

for key in Student :
  print(key)
```

```python
for key, value in Student.items():
    print(key, ":", value)
```

```python
Student = {"Name" : "Tajkir", "Age" : 23}

Student_Copy = Student.copy()                     # Copying Dictionary

print(Student_Copy)
```

```python
Student = {"Name" : "Tajkir", "Age" : 23}

Student_Copy = dict(Student)                      # Using  dict()  function

print(Student_Copy)
```

```python
Company = {                                       # Nested Dictionary
    "HR": {"Manager": "A",  "Staff": 5},
    "IT": {"Manager": "B",  "Staff": 10}
}

print(Company)
```
