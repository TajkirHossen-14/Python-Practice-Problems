```python
# Tuple Slicing                         Tuple[Start : End : Step]

T = (1, 2, 3, 4, 5, 6, 7, 8)            # Print from starting index to just before the last index.

print(f"1.  {T[:]}")
print(f"2.  {T[1:7]}")
print(f"3.  {T[1:8:3]}")
print(f"4.  {T[:4]}")
print(f"5.  {T[4:]}")
print(f"6.  {T[::2]}")
print(f"7.  {T[::-1]}")
print(f"8.  {T[-3:]}")
print(f"9.  {T[-4:-1]}")
print(f"10. {T[-5:4]}")
print(f"11. {T[5:-4]}")
print(f"12. {T[-1:-8:-3]}")
print(f"13. {T[1:100]}")
```

**- Sets -**

```python
S = {1,2,2,3,4}

print(S)
print(type(S))
```

```python
Players = {'Mustafiz','Rishad','Tanzid'}

Players.add('Tajkir')                             # Add Single Element
print(Players)
```

```python
# Mathematical Set Operations

A = {1,2,3,4}
B = {3,4,5,6}

print('Union :', A|B)
print('Intersection :', A & B)
print('Difference :', A-B)
print('Symmetric Difference :', A^B)
```

```python
A = {1, 2}
B = {1, 2, 3, 4}

print("Subset:", A.issubset(B))              # Subset : All Elements of A Are Present Inside B
print("Superset:", B.issuperset(A))          # Superset : B Contains All Elements of A
print("Disjoint:", A.isdisjoint({5, 6}))     # Disjoint : Two Sets Have No Common Elements
```

```python
FS = frozenset([1, 2, 3])          # Frozen Set is Immutable

print(FS)

# FS.add(5)                        # ❌ Cannot Modify  [ERROR]
```

```python
# Frozen Set as Dictionary Key
D = {
    frozenset([1,2]): "Group A",
    frozenset([3,4]): "Group B"
}

print(D)
```

```python
List = [1,2,2,3,3,4,5,5]

Unique = list(set(List))          # Remove Duplicate Elements
print(Unique)
```

```python
# Set Comprehension - Syntax : { Expression for Item in Iterable if Condition }

S = set()
for i in range(5):
    S.add(i)                 # Normal Way of Creating Set (Using Loop)

print(S)

print(".")

T = {x for x in range(1, 7)}
print(T)

print(".")

U = {x for x in range(20) if x % 2 == 0}
print(U)

print(".")

List = [1,2,2,3,3,4,5,5]
Unique = {x for x in List}
print(Unique)
```

```python
L = list(range(1000000))
S = set(L)

print(999999 in L)
print(999999 in S)
```

```python
# Typecasting Various Data Types into Sets

Text = "Bangladesh"                 # String → Set
S1 = set(Text)
print(S1)

print("\n")

L = [1, 2, 2, 3, 4, 4, 5]           # List → Set
S2 = set(L)
print(S2)

print("\n")

D = {"a": 1, "b": 2, "c": 3}       # Dictionary → Set [Only keys are converted into set.]
S3 = set(D)
print(S3)

print("\n")

S4 = {1, 2, 3}          # Set → Set (Copying a Set)
S5 = set(S4)            # Check Whether this is shallow copy or not
print(S5)
```

```python
# Sample Example : Student Registration System

Students = set()

while True:
    S_ID = input("Enter Student ID (or Q to Quit): ")
    if S_ID == 'Q':
        break
    Students.add(S_ID)

print("Registered Students:", Students)
```
